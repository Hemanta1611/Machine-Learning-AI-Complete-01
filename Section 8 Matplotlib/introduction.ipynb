## Matplotlib example workflow

```python
# 0. import matplot lib and get it ready for plotting
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [10, 30, 20, 40]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y);

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis");

# 5. Display plot
plt.show()

# 6. Save plot
fig.savefig("simple-plot-1.png")
```

## Making figures with Numpy arrays

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
# ax.scatter(x, np.tan(x));
```

```python
# Make a plot from dictionary: barplot
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="X's Nut Butter Store", ylabel="Price (₹) -------->", xlabel="Different butter ------->");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# HIstogram: 
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# subplots action 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different axis
ax1.plot(x, x/3);
ax2.scatter(np.random.random(21), np.random.random(21));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(10000));
```

```python
# subplots action 2:
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different index
ax[0, 0].plot(x, x/3);
ax[0, 1].scatter(np.random.random(21), np.random.random(21));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(10000));
```

## PLotting from pandas dataframe

```python
import pandas as pd
```

```python
# Make a data frame
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index = pd.date_range("28-05-2025", periods=1000))
ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('$', '').str.replace(',', '').astype(float).astype(int)
car_sales
```

```python
car_sales["Sale_Date"] = pd.date_range("28-05-2025", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# lets plot the total sale 
car_sales.plot(x="Sale_Date", y="Total Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="bar");
```

```python
# Lets try another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(figsize=(10, 20), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# subplot chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

# add to ax0:
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax0.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle="-.");

# add to ax1:
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="-.");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# subplot chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

# add to ax0:
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="spring")
ax0.set_xlim([50, 80])
ax0.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle="-.");

# add to ax1:
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="rainbow")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="-.");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
